```python
# Every pixel has a color, which is defined by 3 channels(RGB)
# We can use scikit image library to open and close some files using IO.

# Let's import it:

from skimage import io
my_image = io.imread(r"C:\Users\aslan\Downloads\Image Processing\indir.jpg")

# To show the numeric form of the image : 

print(my_image)
```

```python
print(my_image.shape)
```

```python
# According to previous result(output), '3' represents the RGB Channel.

# Let's create an artificial image using numpy library.
# Let's import numpy library:

import numpy as np

# To show images that we created, we can use the matplotlib library...
import matplotlib.pyplot as plt

# Let's create an image with random values...

random_image = np.random.random([500,500])
plt.imshow(random_image)
```

```python
# Let's show the out real image...
plt.imshow(my_image)
```

```python
# For floating point, we can use the img_as_float tool.

from skimage import img_as_float as iaf

my_floating_image = iaf(my_image)

print(my_floating_image)
```

```python
plt.imshow(my_floating_image)
```

```python
# Reducing all images to obtain dark image, we can multiply all values by 0.5:

dark_image = my_floating_image * 0.5
plt.imshow(dark_image)
```

```python
# To fill in the specific region as a particular colour, we can defined code as to give determined coordinates of images:

specific_image = my_image
specific_image[50:75,:50,:] = (0,0,255)
plt.imshow(specific_image)
```
